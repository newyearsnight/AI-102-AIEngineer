```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def read_file(filename):
    with open(filename, 'r') as file:
        text = file.read()
    return text

def eliminate_repeats_and_stopwords(text):
    words = text.split()
    unique_words = set(words)
    stopwords = ["office", "work", "working"]
    filtered_words = [word for word in unique_words if word.lower() not in stopwords]
    return " ".join(filtered_words)

def generate_wordcloud(text):
    wordcloud = WordCloud().generate(text)
    return wordcloud

def display_and_save_wordcloud(wordcloud):
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    wordcloud.to_file("wordcloud.png")

# Read the text file
text = read_file('hybridwork.txt')

# Eliminate repeat words and remove specific words
filtered_text = eliminate_repeats_and_stopwords(text)

# Generate the word cloud
wordcloud = generate_wordcloud(filtered_text)

# Display the word cloud and save it to a file
display_and_save_wordcloud(wordcloud)
```
